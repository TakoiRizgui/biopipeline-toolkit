## 1ï¸âƒ£ Import des Librairies

```python
# BioPipeline Toolkit
from biopipeline.genome.stats import GenomeStats

# Librairies standard
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from datetime import datetime

# Configuration plots
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline

print("âœ… Librairies importÃ©es avec succÃ¨s")
```

## 2ï¸âƒ£ Informations du GÃ©nome

```python
# ===== Ã€ PERSONNALISER =====
GENOME_FILE = "path/to/your/genome.fasta"  # âš ï¸ Modifier ce chemin
GENOME_NAME = "GENOME_001"                 # âš ï¸ Modifier ce nom
ORGANISM = "Bacillus sp."                  # âš ï¸ Modifier l'organisme
SOURCE = "Source thermale, Gafsa"          # âš ï¸ Modifier la source
DATE_ANALYSIS = datetime.now().strftime("%Y-%m-%d")
# ==========================

print(f"ðŸ“ GÃ©nome : {GENOME_NAME}")
print(f"ðŸ§¬ Organisme : {ORGANISM}")
print(f"ðŸ“ Source : {SOURCE}")
print(f"ðŸ“… Date : {DATE_ANALYSIS}")
```

## 3ï¸âƒ£ Chargement et Analyse

```python
# Charger le gÃ©nome
print(f"â³ Chargement de {GENOME_FILE}...")
stats = GenomeStats(GENOME_FILE)
print(f"âœ… {len(stats.sequences)} sÃ©quences chargÃ©es")
```

## 4ï¸âƒ£ Statistiques GÃ©nÃ©rales

```python
# Obtenir toutes les statistiques
basic_stats = stats.get_basic_stats()

# Afficher
print(stats)

# DataFrame pour export
df_stats = pd.DataFrame([basic_stats])
df_stats
```

## 5ï¸âƒ£ Ã‰valuation de la QualitÃ©

```python
# CritÃ¨res de qualitÃ©
n50 = basic_stats['n50']
num_contigs = basic_stats['total_sequences']

# Ã‰valuation
if n50 > 50000 and num_contigs < 100:
    quality = "ðŸŸ¢ EXCELLENTE"
    recommendation = "âœ… ProcÃ©der Ã  l'annotation"
elif n50 > 10000 and num_contigs < 500:
    quality = "ðŸŸ¡ MOYENNE"
    recommendation = "âš ï¸ Annotation possible, mais assemblage amÃ©liorable"
else:
    quality = "ðŸ”´ FAIBLE"
    recommendation = "âŒ RecommandÃ© : RÃ©-assembler avec paramÃ¨tres optimisÃ©s"

print(f"\n{'='*60}")
print(f"Ã‰VALUATION QUALITÃ‰ : {quality}")
print(f"{'='*60}")
print(f"N50 : {n50:,} bp")
print(f"Nombre de contigs : {num_contigs:,}")
print(f"\nðŸ’¡ RECOMMANDATION : {recommendation}")
print(f"{'='*60}\n")
```

## 6ï¸âƒ£ Visualisations

```python
# Distribution des longueurs
stats.plot_length_distribution()
```

```python
# Distribution GC
stats.plot_gc_distribution()
```

## 7ï¸âƒ£ Export des RÃ©sultats

```python
# CrÃ©er dossier rÃ©sultats
output_dir = Path(f"results/{GENOME_NAME}_QC")
output_dir.mkdir(parents=True, exist_ok=True)

# Sauvegarder statistiques
df_stats.to_csv(output_dir / f"{GENOME_NAME}_stats.csv", index=False)
print(f"âœ… Stats sauvegardÃ©es : {output_dir}/{GENOME_NAME}_stats.csv")

# Sauvegarder graphiques
stats.plot_length_distribution(output_dir / f"{GENOME_NAME}_length_dist.png")
stats.plot_gc_distribution(output_dir / f"{GENOME_NAME}_gc_dist.png")
print(f"âœ… Graphiques sauvegardÃ©s dans {output_dir}/")
```
